```python
import random
l=["rock","paper","scissor"]

while True:
    ccount=0
    ucount=0
    start=int(input("""
\U0001F64F Welcome to the game \U0001F64F

\nDo you want to start the game \U0001F4AA :-
1. Yes 
2. No 
    
"""))
    
    if start==1:
        for a in range (1,6):
            User_input=int(input("""
\nSelect from the following :- 
1. rock
2. paper
3. scissor           
            """))
            
            if User_input==1:
                user_choice="rock"
                
            elif User_input==2:
                user_choice="paper"
                
            elif User_input==3:
                user_choice="scissor"
                
            comp_choice=random.choice(l)
            print("Your choice is :- ",user_choice)
            print("Computer choice is :- ",comp_choice)
            
            if user_choice==comp_choice:
                print("Oops, Game draw.")
                ucount=ucount+1
                ccount=ccount+1
            elif (user_choice=="rock" and comp_choice=="scissor") or (user_choice=="scissor" and comp_choice=="paper") or (user_choice=="paper" and comp_choice=="rock"):
                print("Yeah, you win !!!!")
                ucount=ucount+1
            else:
                print("Oops, computer win !!!!")
                ccount=ccount+1
        if ucount==ccount:
            print("\nThe game draw between you and computer.")
            print("Your score :- ",ucount)
            print("Computer score :- ",ccount)
            
        elif ucount>ccount:
            print("\nYeah, you win the game. CHAL ATA PARTY DE...!!!")
            print("Your score :- ",ucount)
            print("Computer score :- ",ccount)
        
        else:
            print("\nOops you lost the game buddy, better luck next time.")
            print("Your score :- ",ucount)
            print("Computer score :- ",ccount)
    else:
        break
```
